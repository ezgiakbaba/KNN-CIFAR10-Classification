```python
import numpy as np
import matplotlib.pyplot as plt
from keras.datasets import fashion_mnist, cifar10
```

```python
(X_train, y_train), (X_test, y_test) = fashion_mnist.load_data()
```

```python
print("X_train shape:", X_train.shape)
print("y_train shape:", y_train.shape)
print("X_test shape:", X_test.shape)
print("y_test shape:", y_test.shape)
```

```python
print("Görüntü boyutları:", X_train[0].shape)
```

```python
plt.figure(figsize=(10,10))
for i in range(10):
    plt.subplot(1,10,i+1)
    if X_train.shape[-1] == 1:  # Tek kanallı (siyah beyaz) görüntüler için
        plt.imshow(X_train[i], cmap='gray')
    else:  # Renkli görüntüler için
        plt.imshow(X_train[i])
    plt.axis('off')
plt.show()
```

```python
X_train = X_train.astype('float32') / 255.0
X_test = X_test.astype('float32') / 255.0
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report

knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train.reshape(X_train.shape[0], -1), y_train)
y_pred = knn.predict(X_test.reshape(X_test.shape[0], -1))

print("Accuracy:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
```
